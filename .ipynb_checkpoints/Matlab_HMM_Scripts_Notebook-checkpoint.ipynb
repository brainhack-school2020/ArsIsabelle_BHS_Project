## 1) Hidden Markov Model (HMM) and Auditory Signal

"Machine learning models are trained to develop broad mathematical representations that capture the variability of target patterns in the input[here FFRs]. These representations are then used to decode target patterns from novel input signals. An important character-istic of machine learning approaches is that they incorporate the modeling of complex distributional input properties (Mohri et al., 2012)." (Llanos et al., 2017)

HMM is widely used with auditory signal, such as speech, because it can handle time-varying auditory signal. It is also indicated with FFR as it can "decode patterns at moderate signal-to-noise ratios" (Llanos et al., 2017)

For more information about the HMM, we suggest the following publication by [Llanos et al., 2017](https://doi.org/10.1016/j.jneumeth.2017.08.010). Consult as well the 'References Section of our '[README.md](https://github.com/brainhack-school2020/ArsIsabelle_BHS_Project/blob/master/README.md) document.

The goal of this Hidden Markov Model Exercise is to __examine if the classifier can accuratly differenciate two very similar FFRs__, here a speech sound |dah| and piano tone with the same fundamental frequency of 98Hz and duration of 100ms, and __evaluate the decoding accuracy__ of HMM with various trials sample size. ## 4) Note on Data Format, File Names and BIDS-EEG

This classifier requires EEG data from each subject separatly as a MATLAB structure array with two fields: 


As recommended by BIDS-EEG, the EEG-FFR data is in .egg format and each datafiles represent the data of one subject for one session. As the classifier generates an output with file names that keep track of the number of stages, the overall number of trials used for ML procedures, the number of trials for averaging, other information related to the subject on the data file names have been codified numerically. This allow for the output generated by the HMM to have a lenght that more easily readable and helps manipulations.   

Thus, there are 26 data files each starting with a 'S' (for subject). The group membership has been indicated with the first number (in the hundred position) such that the group of 13 musicians are coded as S101, S102, S103 etc. and the group of 13 non-musicians is coded as S201, S202, S203, etc.

## Run the HMM

Go to the folder "hmm f0 syllable vs tone"

```matlab
cd hmm f0 syllable vs tone
```

Run the script main.m (as shown in the next cell)

```matlab
% Basics

    close all;
    clear all; 

    p = gcp('nocreate');
    if isempty(p)
    parpool(2);                                                               
    end
    
    delete('results/*.mat')

    rng('shuffle');                                                            
    params = setup; 
    
    totsizes = [50 100 200 500 1000 1500 2000 2300]; %These are the different sample size the HMM will use
    avesizes = round(totsizes/4); % The Average sample size for the test K-fold
    
% Main

    for k=1:length(totsizes)
        
        tic
        
        params.totsize = totsizes(k);
        params.avesize = avesizes(k);
      
        subList = dir([params.ddir,'*.mat']);                                

        for i=1:length(subList)  

            % load subject ffrs

            sub = subList(i).name;
            load([params.ddir,sub]); 
            clsList = intersect(cls,cls);

            % train/test kfolds

            input = struct('ffrs',ffrs,'cls',cls);
            parfor j=1:length(clsList) 
            clsid = clsList(j);
            [trainfolds{j},testfolds{j}] = create_kfolds(input,clsid,params);
            end

            % move-average train folds

            parfor j=1:length(clsList)        
            trainfolds{j} = moving_average(trainfolds{j},params);           
            end

            % f0s

             for j=1:length(clsList)                                              
             trainfolds{j} = ffr2f0(trainfolds{j},params);                               
             testfolds{j} = ffr2f0(testfolds{j},params);                               
             end

            % rearrange kfolds

            [trainfolds,testfolds] = rearrange_kfolds(trainfolds,testfolds);

            % cross-validation

            pcls = [];
            rcls = [];
            plogps = [];

            for j=1:length(testfolds)

                trainfold = trainfolds{j};
                testfold = testfolds{j};
                [right pred logp] = hmm_modeling(trainfold,testfold,params);
                rcls = [rcls;right];
                pcls = [pcls;pred];
                plogps = [plogps;logp];

            end
            
            cm = confusionmat(rcls,pcls);
            sub
            cm = 100*cm/mean(sum(cm,1))

            % save results, clear vars, toc

            outname = [params.rdir,'sub',sub(1:end-4),'_nst',num2str(params.nst),...
            '_size',num2str(params.totsize),...
            '_ave',num2str(params.avesize)];   
            save(outname,'pcls','rcls','plogps');        
            clear trainfolds testfolds

        end
        
        toc
    
    end

```

# Results

The figure below shows the results for overall HMM decoding accuracy of the syllable vs. tone FFRs. Recognition above chance start at about 100 trials (left panel). Differences related to music experience do not
emerge until 1000 trials (center panel). As well, differences related to music experience only emerge for the HMM decoding of the tone (musicians>non musicians); they are not present for the decoding of the syllable stimuli (right panel). 

The reason why music experience differences in overall decoding ( syllable&tone do no emerge even for 1000 trials may be related to the fact that both populations are familiar with the syllable. This is supported by the fact that no music experience differences seem to emerge from the HMM decoding of the syllable (right panel). Therefore, music experience differences seem to be related to the stimulus signal that is more popular among musicians (the tone). 

The reason why it takes more than 100 trials to reach accuracy over chance may be related to the fact that both
the tone and the syllable share the same F0 contour. In this case, it may take hundreds of trials for small,
context determined perturbations to emerge in the FFR.
